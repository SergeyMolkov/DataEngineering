### 1.5 Task_4 Using CTE find users who listened to music in Jan and Feb, then show only users who didn't listen to music in Feb.

```python
sql = ''' with users_jan as (
          select *
          from "Sognify".plays t
          where play_date between '2017-01-01' and '2017-01-31'),

          users_feb as (
          select *
          from "Sognify".plays t
          where play_date between '2017-02-01' and '2017-02-28')

          select t.user_id
          from users_jan t
          left join users_feb ub on t.user_id = ub.user_id
          where ub.user_id is null
          '''
```

```python
select(sql)
```

### 1.6 Task_5 Add active or non-active status for premium users for each month into the months

```python
sql = ''' select t.user_id,
          case when t.purchase_date <= m.months and (t.cancel_date >= m.months or t.cancel_date is null) then 'active' else 'non-active' end as status
          from "Sognify".premium_users t
          cross join "Sognify".months m
          '''
```

```python
select(sql)
```

### 1.7 Task_6 How many times has each song been played

```python
sql = ''' select t.song_id, s.title, s.artist, count(*)
          from "Sognify".plays t
          left join "Sognify".songs s on t.song_id = s.id
          group by t.song_id, s.title, s.artist
          order by t.song_id
          '''
```

### 3.7 Task_6 Which personality type is most common among employees and which projects were chosen by employees with this personality type

```python
sql = ''' with most_popular_personality as (
          select t.personality, count(*) as count_personality
          from "vr_startup".employees t
          group by 1
          order by 2 desc
          limit 1
          )

          select p.project_name, count(*) as count_project
          from "vr_startup".employees t
          join "vr_startup".projects p on t.current_project = p.project_id
          where t.personality = (select t.personality from most_popular_personality t) and t.current_project is not NULL
          group by p.project_name
          order by count(*) desc
          '''
```

```python
select(sql)
```

### 3.8 Task_7 Which personality type is most common among in-project employees and which projects were chosen by employees with this personality type

```python
sql = ''' with most_popular_personality as (
          select t.personality, count(*) as count_personality
          from "vr_startup".employees t
          where t.current_project is not NULL
          group by 1
          order by 2 desc
          limit 1
          )

          select t.first_name, t.last_name, t.personality, p.project_name
          from "vr_startup".employees t
          join "vr_startup".projects p on t.current_project = p.project_id
          where t.personality = (select t.personality from most_popular_personality t)
          '''
```

```python
select(sql)
```

### 3.9 Task_8 Show first name, last name, personality and project name each employee and number of employees with the same personality

```python
sql ='''with count_same_personality as (
        select t.employee_id, count(*)
        from "vr_startup".employees t
        cross join "vr_startup".employees e
        where t.personality = e.personality
        group by 1
        order by 1
        )

        select t.first_name, t.last_name, t.personality, p.project_name, cs.count as number_of_employees_with_the_same_personality
        from "vr_startup".employees t
        left join "vr_startup".projects p on t.current_project = p.project_id
        left join count_same_personality cs on t.employee_id = cs.employee_id
        '''
```

```python
select(sql)
```
